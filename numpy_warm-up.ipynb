**Q9** - Matrisler`G` ve `H` â€™nin toplamÄ±nÄ± hesapla ve sonucu `gh_sum` adlÄ± bir deÄŸiÅŸkende sakla:
$
G = 
\begin{pmatrix}
0 & 4\\

1 & 1
\end{pmatrix}
$

$
H =
\begin{pmatrix}
0 & 1\\
1 & -1\\
2 & 3
\end{pmatrix}
$

$
G + H = ?
$


````python
# Here we demonstrate two different approaches to creating a (3,2) Matrix
G = np.array([0, 4, -4, -3, 1, 1]).reshape(3, 2)
print('G: \n', G)
H = np.array([[0, 1], [1, -1], [2, 3]])
print('H: \n', H)
````

````python
gh_sum = G + H
gh_sum
````
